```python
import requests
url='https://newsapi.org/v2/everything?q=ai&apiKey=da89a6771c3b491fb8d4dfe322a07a01'
params={
    "q": "ai",
    "language": "en",
    "sortBy": "publishedAt",
    "pageSize": 100,
    "page": 1,
    "apiKey": "da89a6771c3b491fb8d4dfe322a07a01"
}
response=requests.get(url,params=params)
print(response.json())
```

```python
articles=response.json()['articles']
for article in articles:
    print(article['title'])
```

```python
import pandas as pd
df=pd.DataFrame(articles)
df
```

```python
df = df[["title","description","content","publishedAt","author"]].copy()
df.head(50)
```

```python
df = df.dropna(subset=["title"])
df["description"] = df["description"].fillna("")
df.drop_duplicates(subset="title")
```

```python
df["publishedAt"] = pd.to_datetime(df["publishedAt"], errors="coerce")
df.head()
```

```python
df["full_text"] = df["title"] + " " + df["description"]
df.head()
```

```python
positive_words = ["good", "growth", "profit", "rise", "success", "gain"]
negative_words = ["loss", "fall", "crash", "fraud", "decline", "drop"]
```

```python
def get_sentiment(text):
    pos = sum(word in text for word in positive_words)
    neg = sum(word in text for word in negative_words)

    score = pos - neg

    if score > 0:
        return "positive", score
    elif score < 0:
        return "negative", score
    else:
        return "neutral", score
```

```python
df[["sentiment", "score"]] = df["full_text"].apply(
    lambda x: pd.Series(get_sentiment(x))
)
```
